# **Most of the buyers are from Maharashtra State**

```python
df["Date"] = pd.to_datetime(df["Date"], format="%m-%d-%y", errors='coerce')  # Convert to datetime
df = df.dropna(subset=["Date"])  # Drop invalid dates
df["Month"] = df["Date"].dt.to_period("M")  # Extract month
df["ship-postal-code"] = df["ship-postal-code"].astype("Int64").astype("string")  # Postal code cleanup
df = df[df["Amount"].notna()]  # Drop rows without Amount
```

```python
monthly_sales = df.groupby("Month").agg(
    Total_Orders=("Order ID", "nunique"),
    Total_Quantity=("Qty", "sum"),
    Total_Revenue=("Amount", "sum")
).reset_index()
monthly_sales["Month"] = monthly_sales["Month"].dt.to_timestamp()
```

```python
sns.set_theme(style="whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_sales, x="Month", y="Total_Revenue", marker="o", color="blue")
plt.title("Monthly Revenue Trend")
plt.ylabel("Total Revenue (INR)")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
fulfillment_stats = df.groupby("Fulfilment").agg(
    Orders=("Order ID", "nunique"),
    Revenue=("Amount", "sum")
).reset_index()
```

```python
plt.figure(figsize=(8, 5))
Fulfilment = sns.barplot(data=fulfillment_stats, x="Fulfilment", y="Revenue", palette="coolwarm")
for bar in Fulfilment.containers:
    Fulfilment.bar_label(bar, fmt='%.2f')
plt.title("Revenue by Fulfillment Method")
plt.ylabel("Total Revenue (INR)")
plt.tight_layout()
plt.show()
```

# **Amazon Sales Data Analysis – Key Insights**

**1. Revenue by Fulfillment Method**

Amazon Fulfillment generated significantly more revenue ~54.3 million INR compared to Merchant Fulfillment ~24.3 million INR.

**Conclusion:** Partnering with Amazon’s fulfillment services drives higher revenue, likely due to better logistics and trust.

**2. Monthly Revenue Trend**

Revenue increased sharply from March to April 2022, peaking in April (~30 million INR).

Then it slightly decreased in May and June.

**Conclusion:** April was the best performing month. This could relate to promotional events, seasonality, or marketing campaigns.

**3. Top 10 Shipping States**

Maharashtra leads by a large margin (22,272 shipments), followed by Karnataka (17,324) and Tamil Nadu (11,488).

**Conclusion:** Focus marketing and logistics efforts on high-volume states like Maharashtra and Karnataka for better ROI.

**4. Category vs. Size Distribution**

A wide variety of product sizes (from XS to 6XL and Free Size) are available across multiple categories like T-shirts, Shirts, Shoes, Wallets, etc.

**Conclusion:** Inventory is well-diversified in sizes, supporting a broad customer base with different needs.

**5. B2B vs B2C Customers**

99.3% of buyers are Retail (B2C), while only 0.7% are B2B (wholesale).

**Conclusion:** The customer base is primarily individual retail buyers. There's potential to expand into B2B sales.

**6. Product Category Distribution**

T-shirts and Shirts dominate the product categories, each with ~50,000 items sold.

Blazers and Trousers follow with much lower volumes.

**Conclusion:** Focus on optimizing T-shirt and Shirt inventory and explore marketing strategies to boost lower-selling categories.

**7. Courier Status Distribution**

Most orders are marked as “Shipped - Delivered to Buyer” (77,607) or “Shipped - Returned to Seller” (28,748).

Other statuses (like damaged, cancelled, lost in transit) are very minimal.

**Conclusion:** Delivery efficiency is high, but there is room to reduce return rates.

**8. Size Distribution**

The most common sizes are M (22,373), L (21,807), and XL (20,549).

Rare sizes like 4XL, 5XL, XS have much lower counts.

**Conclusion:** Inventory planning should prioritize mid-range sizes (M–XL), which cover the majority of the demand.
